```
# 데이터를 추가할 빈 그릇을 하나 만든다.

data_petitions_contents = pd.DataFrame()


# HTML 받아오기
  
page = "https://spacecloud.kr/search?q=%EC%BD%94%EC%9B%8C%ED%82%B9+%EC%8A%A4%ED%8E%98%EC%9D%B4%EC%8A%A4"
 
with request.urlopen(page) as f:
  charset = f.headers.get_content_charset()
  html = f.read().decode(charset)
   
# HTML 문자열을 분석하여 DOM 구성하기

bs_coworkspace = BeautifulSoup(html, 'html5lib')

# 각 페이지의 제목, 본문 데이터를 뽑아 데이터프레임에 추가한다.

coworkspace = bs_coworkspace.select('.info_area')
  

coworkspace
```
